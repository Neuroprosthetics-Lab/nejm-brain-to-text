```python
import os
import re
import time
import datasets

from tqdm import tqdm
```

## Load HF Dataset

```python
# Load the lite configuration of the dataset
raw_dataset = datasets.load_dataset("JanosAudran/financial-reports-sec", "large_lite")
```

```python
with open(os.path.join('corpus', 'financial-reports-sec.txt'), 'w') as fw:
    for split in ['train', 'validation', 'test']:
        for sample in tqdm(raw_dataset[split]):
            fw.write(sample['sentence'] + '\n')
```

## Run ngram script

```python
LM_ROOT_DIR = os.path.abspath('../examples/speech/s0')
RUN_SCRIPT_PATH = os.path.join(LM_ROOT_DIR, 'run.sh')
DICT_PATH = os.path.join(LM_ROOT_DIR, 'dict.txt')
CORPUS_ROOT_DIR = os.path.abspath('../../../ngram-lm/corpus')
GENERATED_LM_ROOT_DIR = os.path.abspath('../../../ngram-lm/generated_lm')

CORPUS_FILENAME = 'financial-reports-sec_processed.txt' #'financial-reports-sec.txt'
LM_NAME = 'financial_reports_{ORDER}gram_test'

CORPUS_PATH = os.path.join(CORPUS_ROOT_DIR, CORPUS_FILENAME)
FORMATTED_CORPUS_PATH = 'none'
SIL_PROB = 0.9
PRUNE_THRESHOLD = 0
ORDER = 3
OUTPUT_PATH = os.path.join(GENERATED_LM_ROOT_DIR, LM_NAME.format(ORDER=ORDER))

os.makedirs(OUTPUT_PATH, exist_ok=True)
```

```python
command = f'cd {LM_ROOT_DIR} && bash {RUN_SCRIPT_PATH} {OUTPUT_PATH} {DICT_PATH} {CORPUS_PATH} {SIL_PROB} {FORMATTED_CORPUS_PATH} {PRUNE_THRESHOLD} {ORDER}'
command
```
